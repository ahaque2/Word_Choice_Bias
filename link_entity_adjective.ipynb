```python

def get_spacy_ner(text):

    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)
    ner = {}
    for ent in doc.ents: 
        if ent.label_ in ner:
            ner[ent.label_].append(ent.text)
        else:
            ner[ent.label_] = [ent.text]
        
    for key in ner:
        ner[key] = set(ner[key])
        
    return ner

def get_entity_adj_pairs(spacy_nlp, sent):
    
    extracted_adj = dict()
    
    doc = spacy_nlp(sent)
    #print(doc.ents)
    #sys.exit()
    ner = {}
    for ent in doc.ents:
        if(ent.label_ == 'PERSON'):
            #print("sent ", sent)
            #print("entity: ", ent)
            for tok in doc:
                if tok.tag_ == 'JJ':
                    extracted_adj[ent] = tok.lemma_
                    #print(tok.lemma_)
    return extracted_adj

def get_details_of_occurence_of_words(text, word):
    
    identified_sent = dict()
    if(word in text):
        sent = sent_tokenize(text)
        for i, s in enumerate(sent):
            if word in s:
                identified_sent[i] = s.encode('ascii', errors='ignore').decode("utf-8")
                
    return identified_sent
                
        
    
```
